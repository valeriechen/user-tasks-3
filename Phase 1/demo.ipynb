```python
import matplotlib.pyplot as plt
from toggle_dark_mode import toggle_dark_mode

# Create a sample plot
fig, ax = plt.subplots()
ax.plot([0, 1, 2, 3], [10, 20, 25, 30], label="Sample Line")
ax.set_title("Sample Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.legend()

# Show the plot in light mode
plt.show()
```

```python
print(fig is not None)
# Toggle to dark mode
toggle_dark_mode(fig=fig)

# Show the plot in dark mode
display(fig)
```

```python
# Toggle back to light mode
toggle_dark_mode(fig=fig)

# Show the plot in light mode again
display(fig)
```
